# Crop size dependence

```python
model_id = 0
model_name,W_model = model_list[model_id]
print(f'{model_name=}')
layer_names = layers_dict[model_name][:] 
class_list = list(class_dict.keys())[:]

layer_ids = [0,2,3,5,6]
_relative_depth_dict = [relative_depth_dict[model_name][idx] for idx in layer_ids]
class_ids = [0,1,2,3,4,5,6]
crop_sizes = 28 * np.arange(1,8+1,dtype=int)

shuffle_flags = [0,1]
IDs = np.zeros(shape=(2, # (GRIDE, FCI, BID)
                      len(layer_ids),
                      len(class_ids),
                      len(crop_sizes),
                      len(shuffle_flags))
                      )

                      

for shuffle_id,shuffle in enumerate(shuffle_flags):
  for index_l,layer_id in enumerate(layer_ids):
    for index_c,class_id in enumerate(class_ids):
      for index_crop,crop_size in enumerate(crop_sizes):
        layer_name = layer_names[layer_id]
        key = class_list[class_id]
        ### GRIDE
        base = f'GRIDE/results/gride/'
        if shuffle:
          base += 'shuffle/'
          key=class_id
        resultsfolder = makefolder(base=base,
                              create_folder=False,
                              crop_size=crop_size,
                              key=key,
                              layer_id=layer_id,
                              )
        filename='gride.txt'
        (ids_gride,ids_err_gride,rs_gride) = np.loadtxt(resultsfolder+filename,unpack=True)
        selected_scale_id = np.argmin(np.abs(np.diff(ids_gride)))
        selected_scale_id = 3
        IDs[0,index_l,index_c,index_crop,shuffle_id] = ids_gride[selected_scale_id]
        ### FCI
        base = f'FCI/results/FCI/'
        if shuffle:
          base += 'shuffle/'
          key=class_id
        resultsfolder = makefolder(base=base,
                              create_folder=False,
                              crop_size=crop_size,
                              key=key,
                              layer_id=layer_id,
                              )
        filename='FCI.txt'
        (d,x0,err) = np.loadtxt(resultsfolder+filename,unpack=True)
        IDs[1,index_l,index_c,index_crop,shuffle_id] = d   
```

```python
layer_index = 1
layer_name = layer_names[layer_ids[layer_index]]

### BID
BIDfolder = f'BID/crop_size_dependence/results/'
delta = 5E-4
Nsteps = int(1E6)
seed = 1
alphamax = 0.5
scrop_sizes,smu,sstd = np.loadtxt(fname=f'{BIDfolder}patches/{layer_name}_alphamax{alphamax:.5f}.txt',unpack=True)  
bid_crop_sizes,bid_mu,std = np.loadtxt(fname=f'{BIDfolder}ImageNet/{layer_name}_alphamax{alphamax:.5f}.txt',unpack=True)  
```

```python
fig,axs = plt.subplots(1,2,figsize=(10,5))

mean_IDs = np.mean(IDs,axis=2)
std_IDs = np.std(IDs,axis=2)

for shuffle_id, shuffle in enumerate(shuffle_flags):
  if shuffle:
    title = r'$l/L_R=$' + f'{_relative_depth_dict[layer_index]:.2f}'
  axs[shuffle_id].plot(crop_sizes**2 / crop_sizes[-1]**2,
              mean_IDs[0,layer_index,:,shuffle_id],
              # std_IDs[layer_index,:,shuffle_id],
              # fmt='-o',
              'o-',
              label='GRIDE',
              # color=colors[shuffle],
              )
  axs[shuffle_id].plot(crop_sizes**2 / crop_sizes[-1]**2,
              mean_IDs[1,layer_index,:,shuffle_id],
              # std_IDs[layer_index,:,shuffle_id],
              # fmt='-o',
              'o-',
              label='FCI',
              # color=colors[shuffle],
              )
axs[1].plot(scrop_sizes,smu,'o-',label='BID')
axs[0].plot(bid_crop_sizes,bid_mu,'o-',label='BID')

box = axs[1].get_position()
axs[1].set_position([box.x0, box.y0, box.width * 0.8, box.height])
axs[1].legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
axs[1].set_xlabel(r'$N_{crop}/N_{tot}$')
axs[0].set_xlabel(r'$N_{crop}/N_{tot}$')
axs[0].set_ylabel(f'ID')


axs[0].set_yscale('log')
axs[1].set_yscale('log')

# axs[0].set_xscale('log')
# axs[1].set_xscale('log')

axs[0].set_ylim(np.min(mean_IDs),np.max(smu)*1.1)
axs[1].set_ylim(np.min(mean_IDs),np.max(smu)*1.1)

plt.suptitle(title)
axs[0].grid()
axs[1].grid()

axs[0].set_title('ImageNet')
axs[1].set_title('patches')
plt.show()
```

```python
import numpy as np
import sys,os
from time import time
import pyFCI
import matplotlib.pyplot as plt

start = time()
geometry = 'Ising-square'
eps = 1E-7
L = 100

T_list = np.arange(1.8,3+eps,.1)
T_list = [2]
ks = [4]

center = 1

ds = []
for T in T_list:
  N = L**2
  datafolder = f'/scratch/sacevedo/{geometry}/canonical/'
  datafile = datafolder + f'L{L}/T{T:.2f}.txt'
  print(f'{T=:.2f}')
  X = np.loadtxt(f'{datafile}').astype(int)[:1000]
  r = pyFCI.local_FCI(X,center,ks)
  ds.append(r[0,-3])
```

```python
plt.plot(T_list,ds)
```

```python
X = pyFCI.center_and_normalize(X)
fci = pyFCI.FCI(X)
d,x0,err = pyFCI.fit_FCI(fci)
```

```python
d
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(fci[:,0], fci[:,1], label="empirical")

xs = np.linspace(0,2,100)
ys = pyFCI.analytical_FCI(xs,d-1,1)
ax.plot(xs, ys, label="model")

ax.legend(loc='upper left')
ax.set_xlim([0,2])
ax.set_ylim([0,1])

plt.plot()
```
